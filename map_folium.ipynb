```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
data = pd.read_csv('miami-housing.csv')
data.head()
```

```python
old_cols = list(data.columns)
new_cols = ['долгота', 'широта', 'уникальный_номер', 'стоимость_дома', 'площадь_дома_футы', 'площадь_земельного_участка', 'площадь дома', 'стоимость_особых_объектов', 'расст_до_ближ_жд', 'расст_до_океана', 'расст_до_ближ_водоема', 'расст_до_делового_района', 'расст_до_ближ_субцентра', 'расст_до_ближ_шоссе', 'возраст_строения', 'уровень_шума', 'месяц_продажи', 'качество_конструкции']
translation = dict(zip(old_cols, new_cols))

data = data.rename(columns=translation)
data.head()
```

```python
dolgota = data['долгота']
shirota = data['широта']
price = data['стоимость_дома']
area = data['площадь_дома_футы']

def color_change(price):
    if(price < 235000.0):
        return('green')
    elif(235000.0 <= price <310000.0):
        return('orange')
    else:
        return('red')
    
# m = folium.Map(location=[25.761681,-80.191788], zoom_start = 15, tiles = "Cartodb Positron")

m = folium.Map(location=[25.761681,-80.191788], zoom_start = 12, tiles = "Cartodb Positron")

for dolgota, shirota, price, area in zip(dolgota, shirota, price, area):
    folium.CircleMarker(location=[dolgota, shirota], radius = 4, tooltip=f"{price} $", popup=f"Площадь дома: {area} (футы), цена: {price} $", fill_color=color_change(price), color="gray", fill_opacity = 0.9).add_to(m)
m.save("map1.html")
```

```python
#перевод из футов в метры

data['площадь_дома_футы'] = data['площадь_дома_футы'].apply(lambda x: x * 0.09290304)
```

```python
# кластеры

dolgota = data['долгота']
shirota = data['широта']
price = data['стоимость_дома']
area = data['площадь_дома_футы']

def color_change(price):
    if(price < 235000.0):
        return('green')
    elif(235000.0 <= price <310000.0):
        return('orange')
    else:
        return('red')
    

m = folium.Map(location=[25.761681,-80.191788], zoom_start = 12, tiles = "Cartodb Positron")

marker_cluster = MarkerCluster().add_to(m)
for dolgota, shirota, price, area in zip(dolgota, shirota, price, area):
    folium.CircleMarker(location=[dolgota, shirota], radius = 5, tooltip=f"{price} $", popup=f"Стоимость: {price} $, площадь дома: {area} кв. м", fill_color=color_change(price), color="gray", fill_opacity = 0.9).add_to(marker_cluster)

m.save("map3.html")
```
